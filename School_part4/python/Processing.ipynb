```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('xiaoxue.csv')
data.info()
data.head()
```

```python
data.isnull().sum()
```

```python
data['Price'] = data.Price.astype(float)
data['Price'] 
df=data.copy()
```

```python
df_count = df.groupby('School')['Price'].agg(Count='count').reset_index()
# #df_neighborhood_type_count
df_grouped = df.groupby('School')["Price"]
#Calculate the mean, median, maximum and variance of prices across the region
df_price = df_grouped.agg([np.mean]).sort_values("mean")
df_price
# #Merge the dataframe
df_type = pd.merge(df_count,df_price,left_on='School', right_index=True)
df_type
```

```python
df2 = df_type.loc[:,[ 'School', 'mean']]
df2
```

```python
data1 = pd.read_csv('jingweidu.csv')
data1.info()
```

```python
df3 = pd.merge(df2,data1,on='School', how = 'inner')
df3
```

```python
df3.isnull().sum()
```

```python
df3.to_csv('new data set.csv',encoding="utf_8_sig")
```
